```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/content/bottoms.csv')
df.rename(columns={'user/seller_score/rating_average': 'rating'}, inplace=True)
df = df[['price', 'condition', 'rating', 'location']].dropna()

# 1. Bar Plot: Count by condition
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='condition')
plt.title("Number of Items by Condition")
plt.show()

# 2. Boxplot: Price by condition
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='condition', y='price')
plt.title("Price Distribution by Condition")
plt.show()

# 3. Scatterplot: Rating vs. Price
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='rating', y='price', hue='condition')
plt.title("Rating vs Price by Condition")
plt.show()

# 4. Bar: Avg price by location (Top 10 locations)
top_locs = df['location'].value_counts().head(10).index
avg_price = df[df['location'].isin(top_locs)].groupby('location')['price'].mean().sort_values(ascending=False)

plt.figure(figsize=(8,5))
avg_price.plot(kind='bar')
plt.title("Average Price by Top 10 Locations")
plt.ylabel("Avg Price")
plt.xticks(rotation=45)
plt.show()
```
